```python
import numpy as np
import pandas as pd

path = 'RCNN_Scratch'
train_loss = pd.read_pickle(path+'_TrainLoss.pkl')
val_loss = pd.read_pickle(path+'_ValLoss.pkl')
```

```python
from matplotlib import pyplot as plt
loss_col = ['rpn_loc_loss',
            'rpn_cls_loss',
            'roi_loc_loss',
            'roi_cls_loss',
            'total_loss'
            ]
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(30)
epochs = [i for i in range(8)]
i = 1
for loss_type in loss_col:
    tl = train_loss[loss_type]
    vl = val_loss[loss_type]
    plt.subplot(1, 5, i)
    plt.plot(epochs, tl)
    plt.plot(epochs, vl)
    plt.xlabel('Epoch',fontsize=18)
    plt.ylabel('Loss',fontsize=16)
    plt.title(loss_type,fontsize=24)
    plt.legend('train_loss', 'val_loss')
    plt.rc('xtick',labelsize=14)
    plt.rc('ytick',labelsize=14)
    #plt.tight_layout()
    #plt.gca().set_aspect('equal', adjustable='box')
    i += 1
plt.show()
fig.savefig('plot/'+path+'_Loss.png')
```
